```python
data_desc = {}

for col in df:
    if df[col].dtype == 'int64':
        data_desc[col] = 'Cont'
    elif df[col].dtype == 'object':
        data_desc[col] = 'Disc'

print(data_desc)
```

```python
def imputation_nan_values(df, data_desc):
    for col in df:
        if df[col].isnull().values.any():
            if data_desc[col] == 'Disc': 
                df[col].fillna(df[col].mode()[0], inplace=True)
            elif data_desc[col] == 'Cont':
                df[col].fillna(df[col].mean()[0], inplace=True)
```

```python
imputation_nan_values(df, data_desc)
```
